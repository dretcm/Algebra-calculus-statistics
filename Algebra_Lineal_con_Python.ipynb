# Transposición, suma de matrices y escalares

---



```
import numpy as np

matriz = np.array([[1,2],[3,4],[5,6]])

matriz_T = np.zeros(matriz.shape).reshape(matriz.shape[1],matriz.shape[0])

for i in range(matriz.shape[1]):
  for f in range(matriz.shape[0]):
    matriz_T[i][f] = matriz[f][i]

print(matriz)
print(matriz_T)
print(matriz.T)
```

```
matriz2 = np.array([[1,1],[1,1],[1,1]])

res = []
for i in range(len(matriz)):
  res.append([0,0])
print('1 form:',res)

res = []
for i in range(len(matriz)):
  res += [[0,0]]
print('2 form:',res)

res = [ [0,0] for i in range(matriz.shape[0])]
print('3 form:',res)

res = np.zeros(matriz.shape)
print('4 form:',res)
```

```
for i in range(matriz.shape[0]):
  for f in range(matriz.shape[1]):
    res[i][f] = matriz[i][f] + matriz2[i][f]
print('1 sum form: \n',res)

res = matriz + matriz2
print('2 sum form: \n',res)

i = 0
f = 0
while i < res.shape[0]:
  f = 0
  while f < res.shape[1]:
    res[i][f] = matriz[i][f] + matriz2[i][f]
    f +=1
  i +=1
print('3 sum form: \n',res)
```

# Propiedades de las matrices: la multiplicación de matrices es asociativa y distributiva, no es conmutativa

---
**Asociativa:**

A x (B x C) = (A x B) x C

**Distributiva:**

A(B + C) = AB + AC

**Commutativa:**

A x B = B x A

**Nota:**

**El producto interno de matrices es:**

Asociativa: Sí

Distributiva: Sí

Conmutativa: NO

**El producto interno de vectores es:**

Asociativa: Sí

Distributiva: Sí

Conmutativa: Sí

```
import numpy as np

a = np.array([[1,1],[1,2],[3,7]]) #3x2
b = np.array([[2,5],[3,4]]) #2x2
c = np.array([[6,5],[9,1]]) #2x2

# Matriz Asociativa:

print((a.dot(b.dot(c))) == (a.dot(b).dot(c)))
```

```
# Matriz Distributiva:

print((a.dot(b + c)) == (a.dot(b) + a.dot(c)))
```

```
# Matriz Commutativa:

try:
  print(a.dot(b) == b.dot(a))
except ValueError:
  print("\n el producto de matrices no tiene propiedad commutativa")
```

```
# Vector Comutativo:

a = np.array([[4],[5]])
b = np.array([[3],[6]])

print((a.T.dot(b)) == (b.T.dot(a)))
```

# Cómo comprobar la solución de un sistema de ecuaciones lineal

---

y = 3x + 5

y = 2x + 3


-----

3x + 5 = 2x + 3

3x - 2x = 3 - 5

x = -2

----
y = 2x + 3

y = 2(-2) + 3

y = -1

----

y = 3x + 5

y = 2x + 3

y - 3x = 5

y - 2x = 3

---

-3x + y = 5

-2x + y = 3

----

-----A------x------B

| -3   1 | |x|  = |5|

| -2   1 | |y|  = |3|

-3x + 1y = 5

-2x + 1y = 3

----

**Ax = B**

**x = B/A**


```
import numpy as np
import matplotlib.pyplot as plt

A = np.array([[-3,1], [-2,1]]) # matriz A
B = np.array([[5],[3]]) # matriz B
x = np.arange(-5,5)

plt.plot(x,3*x + 5, color = 'orange',label='3x + 5 = y')
plt.plot(x,2*x + 3, color = 'blue',label='2x + 3 = y')
plt.xlim(-5,5)
plt.ylim(-5,5)
plt.axvline(x=0, color = 'grey')
plt.axhline(y=0, color = 'grey')
plt.title('system of equations')
plt.xlabel('eje x')
plt.ylabel('eje y')
plt.legend()
plt.show()
```

```
x = np.array([[-2],[-1]]) # x,y
print(A,'\n X \n',x,'\n = \n',A.dot(x),'\n = \n',B) # Ax = B
```

```
A_inv = np.linalg.inv(A) # A**-1 || 1/A || halla la inversa de una matriz

new_x = A_inv.dot(B) # producto matricial

print(A_inv,'\n X \n',B,'\n = \n',new_x,'\n = \n',x) # A/B = x || (A**-1)(B) = x
```

# Tipos especiales de matrices: Identidad, Inversa, Singulares

---

Matriz identidad = Como el 1 en matematicas, el elemento neutro.

Matriz Inversa = Cumple lo siguiente: A*A^-1 = Id.

Matriz Singular = Cuando No se puede calcular su Inversa, las matrices que son su misma inversa, son involutivas.

¿Porque es importante saber la matriz inversa?

Si tenemos:

A . x = b

donde:

A: Matriz formada por los valores que acompaña a las incognitas.
x: Matriz incognito.
b: Matriz formada por los valores que NO acompañan a las incognitas.
Sabiendo la inversa, podemos calcular los valores de x.

¿Como? Si la A está como productor escalar con la x, pasa al otro lado como producto escalar de b pero como su inversa.

x = b . A^-1

¿Porque es importante saber la matriz singular?

Porque si tenemos una matriz singular es un sistema de ecuaciones lineales no podemos encontrar los valores incongnitas porque no tiene inversa. Sabiendo esto sabemos que no podremos calcular u obtener los resultados de las incognitas.


```
import numpy as np

identidad = np.eye(4)

print(identidad)
```

```
A = np.array([[1,2,3], [0,1,4], [0,0,1]]) # 3x3

A_inv = np.linalg.inv(A) # A**-1

print(A.dot(A_inv))
```

```
singular = np.array([[1,1],[1,1]])
try:
  print(np.linalg.inv(singular))
except np.linalg.LinAlgError:
  print('no tiene matriz inversa')
```

en conclusion: si el determinante de una matriz es “0” es una matriz Singular.

```
# mas ejercicios con la inversa

import numpy as np
# A    x   B
# 3x + y = 1
# 2x + y = 1

A = np.array([[3,1],[2,1]]) 

B = np.array([[1],[1]])

# Ax = B => B/A = x => x = B * A^-1

A_inv = np.linalg.inv(A)

x = A_inv.dot(B)

print(x)
```

```
# Ax = B
print(A.dot(x) == B)
```

# combinación lineal

---


Una combinación lineal de dos o más vectores es el vector que se obtiene al sumar esos vectores multiplicados por escalares. Cualquier vector se puede poner como combinación lineal de otros que tengan distinta dirección. Esta combinación lineal es única.

![link text](https://www.superprof.es/apuntes/wp-content/uploads/2019/07/combinacion-lineal-de-vectores-en-el-espacio-2.gif)


```
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

v1 = np.array([2,5])
v2 = np.array([3,2])

# vecs: vectores.
# cols: colores.
# alpha: valor de transparencia.
def graficarVectores(vecs, cols, alpha=1):
    plt.figure()
    plt.axvline(x=0, color="grey", zorder=0)
    plt.axhline(y=0, color="grey", zorder=0)
    
    for i in range(len(vecs)):
        # El origen de los vectores inicia en el punto (0,0)
        x = np.concatenate([[0,0], vecs[i]])
        plt.quiver([x[0]],
                  [x[1]],
                  [x[2]],
                  [x[3]],
                  angles='xy', scale_units='xy', scale=1, 
                  color=cols[i], alpha=alpha)

graficarVectores([v1,v2], ['orange', 'blue'])
# Definimos los limites
plt.xlim(-1, 8)
plt.ylim(-1, 8)
```

```
v1_c_v2 = 2*v1 + 3*v2

graficarVectores([v1,v2,v1_c_v2], ['orange', 'blue','green'])
# Definimos los limites
plt.xlim(-1, 20)
plt.ylim(-1, 20)
```

# Tipos de normas: norma 0, norma 1, norma 2, norma infinito y norma L2 al cuadrado

---

La norma se usa en machine learning para encontrar errores.

La norma L0: devuelve la cantidad de elementos diferentes de 0 del vector.

La norma L1: devuelve la suma de los valores absolutos de los elementos del 
vector.

La norma L2: La magnitud del vector desde su origen.

La norma (L2)^2: X.dot(X.T)

La norma infinito: devuelve el valor absoluto más grande del vector


**El producto interno como función de una norma y su visualización**

V1.T * V2 = norma(V1) * nroma(V2) * cos(angulo_formado_entre_V1_y_V2) 

```
import numpy as np

x = np.array([1,2,0,0,3,4,4])

# L0

print(np.linalg.norm(x, ord=0))
```

```
# L1
x = np.array([-2,3,5,10])
print(np.linalg.norm(x,ord=1))
```

```
# L2

x = np.array([3,4])

print(np.linalg.norm(x)) # la funcion "norm" tiene por defectuo la norma '2'
print(np.linalg.norm(x,ord=2))
```

# Descomposición de matrices

---
**Descomponer una matriz:** Quiere decir encontrar dos o más matrices que me ayuden a escribir la matriz original y que tengan ciertas propiedades.
Una matriz A la podemos escribir como: sus autovectores producto punto una matriz diagonal, donde la matriz diagonal tiene todos los autovalores encontrados, producto punto la matriz inversa de sus autovectores.

A = AUTOVECTORES * DIAG_AUTOVALORES * INV_AUTOVECTORES

A_calc = autovectores.dot(np.diag(autovalores)).dot(np.linalg.inv(autovectores))

**Descomposición 2:** Restricciones:Nuestra Matriz A sea real y simétrica (A = A-transpuesta). Decimos que A = la matriz de autovectores producto punto la matriz diagonal con los autovalores, producto punto la transpuesta de la matriz de autovectores. NOTA: Calcular la transpuesta de una matriz es mucho más económico en cómputo que calcular la inversa de una matriz. Y hay que recordar que el cómputo cuesta dinero.

A_calc = autovectores.dot(np.diag(autovalores)).dot(autovectores.T)

A = AUTOVECTORES * DIAG_AUTOVALORES * AUTOVECTORES_TRANSPUESTA

![link text](https://static.platzi.com/media/user_upload/descomposicion_matrices_1-a1d5f87a-5432-4fac-b3b9-ffe61de66ac1.jpg)

```
import numpy as np

A = np.array([[3,2],[4,1]])

AutoValores, AutoVectores = np.linalg.eig(A)

DiagAutoValores = np.diag(AutoValores)

InvAutoVectores = np.linalg.inv(AutoVectores)

print(AutoVectores, DiagAutoValores, InvAutoVectores)
```

```
A_calculo = AutoVectores.dot(DiagAutoValores).dot(InvAutoVectores)
print(A,A_calculo)
```

```
# FORMA 2, cuando la matriz sea real y simetrica.

B = np.array([[4,5],[5,10]])

AutoValores, AutoVectores = np.linalg.eig(B)

Dig_AutoValores = np.diag(AutoValores)
T_AutoVectores = AutoVectores.T

print(B,Dig_AutoValores,T_AutoVectores)
```

```
B_Calculus = AutoVectores.dot(Dig_AutoValores).dot(T_AutoVectores) # utilizar la tranpuesta es mas economico computacionalmente.
print(B,B_Calculus)
```

```
import numpy as np
import matplotlib.pyplot as plt

def graficarMatrices(matriz, vector_color=['red','blue']):
  # circulo unitario
  x = np.linspace(-1,1,100000)
  y = np.sqrt(1-(x**2))

  #circulo unitario transformado
  x1 = matriz[0,0]*x + matriz[0,1]*y
  y2 = matriz[1,0]*x + matriz[1,1]*y
  x1_neg = matriz[0,0]*x - matriz[0,1]*y
  y2_neg = matriz[1,0]*x - matriz[1,1]*y

  #vectores
  u1 = [matriz[0,0], matriz[1,0]]
  v1 = [matriz[0,1], matriz[1,1]]

  graficarVectores([u1,v1],vector_color)

  plt.plot(x1,y2,'green',alpha=0.7)
  plt.plot(x1_neg,y2_neg,'green',alpha=0.7)
```

La forma en que python obtiene la pseudo inversa es con el método SVD.

SVD es mas certera, y computacionalmente más efectiva.

La pseudo inversa de Moore Penrose es una aplicación directa de SVD, que nos permite resolver en determinados momentos sistemas de ecuaciones lineales con múltiples soluciones.


Ax = B " y si existe A " x = B/A " o " B * A^-1

A * A^-1 = ID

A * A_pse = ID

A_pse = V.T * D_pse * U.T # utilizaremos V.T en este caso

" ese A_pse" es como decir, A pseudo inversa o algo así y definido como el resipruoco a la inversa de cierta matriz. y para hallar la D_pse, primero coomo ya sé sabe D es la diagonal de la matriz, entonces se sacara la inversa de D, luego se remplazara en una matriz con ceros que tenga el tamaño de la matriz A, donde solo remplazaremos los valores que cuadran en la matriz de cero de dimensiones A.

```
import numpy as np

np.set_printoptions(suppress = True)  #para q no muestre numeros muy cercanos a cero.

A = np.array([[2,3],[5,7],[11,13]])

print(A)
```

```
U, D, V= np.linalg.svd(A)

print("U")
print(U)
print("D")
print(D)
print("V")
print(V)
```

```
D_pse = np.zeros(A.shape).T
print(D_pse.shape)
print(D_pse)
```

```
print('valores a remplazar en D_pse')
print(D_pse[:D.shape[0],:D.shape[0]])
```

```
print('valores que pondremos en D_pse')
D_inv = np.linalg.inv(np.diag(D))
print(D_inv)
```

```
print('D_pse')
D_pse[:D.shape[0],:D.shape[0]] = D_inv
print(D_pse)
```

```
A_pse = V.T.dot(D_pse).dot(U.T)
print('A_pse')
print(A_pse)
```

```
A_pse_calculo = np.linalg.pinv(A) # calculo de la pseudo inversa
print(A_pse_calculo)
```

```
print('aproximacion')
print(A_pse.dot(A))

print('original')
np.set_printoptions(suppress=False)
print(A_pse.dot(A))
```

```
print('no son comutativas')
print(A.dot(A_pse)) # A_pse*A != A*A_pse
```

```
A_pse_2 = np.linalg.inv(A.T.dot(A)).dot(A.T)
print(" similar a la de SVD o penrose") # como se ve es similar a la calculada, pero la de SVD de python es mejor, mejor precision, coste, etc.
print(A_pse_2)
```

*   La maldición de la dimensión: Esto dice que por cada variable que agrego en el conjunto de datos vamos a necesitar exponencialmente más muestras para poder tener la misma relevancia estadística.
*   Cuál es el autovector relacionado con el autovalor más grande? Es el autovalor quien define cual es la dirección que contiene más información.

*   PCA es una técnica muy útil para reducir la cantidad de dimensiones con las que estamos trabajando.
*   Muchas veces nuestro conjunto de datos es muy grande y es necesario reducirlo al menos en un 20%, dejando el 80% de datos mas significativo.


```
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
```

```
x = 3*np.random.rand(200) # hace un reparticion de datos con variacion de 1 o 0
y = 20*x + 2*np.random.randn(200) # random q da numeros crecientes y decrecientes

x = x.reshape(200,1)
y = y.reshape(200,1)

print(x)
print(y)
```

```
xy = np.hstack([x,y])
print(xy)
```

```
plt.plot(xy[:,0],xy[:,1],'.')
plt.show()
```

```
xy_center = xy - np.mean(xy,axis=0)

plt.plot(xy_center[:,0],xy_center[:,1],'.')
plt.show()
```

```
autovalores, autovectores = np.linalg.eig(xy_center.T.dot(xy_center))
print(autovectores) # el vector asociado con el auto valor más grande, nos dice la direccion con maxima variaza.
```
