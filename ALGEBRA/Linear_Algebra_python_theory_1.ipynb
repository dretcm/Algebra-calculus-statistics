**Base:** se le dice base a los vectores unitarios de un sistema de coordenadas. Como (i,j,k),(v,w,u), etc. Que pertenecen a $B$.

por definicion:

La base de un espacio vectorial es un conjunto de vectores lineamente independientes que generan todo el espacio.

**Combinacion lineal:** es la suma de vectores escalados(depende de la dimensionalidad): 

$v = a*i + b*j: a,b  $ ϵ $ R $ & $ i,j $ ϵ $ B$


```
import numpy as np
import matplotlib.pyplot as plt

def draw_vectors(vectors, colors='red', xlim=-1, ylim=5, title=''):
  for i,v in enumerate(vectors):
    try:
      plt.quiver([0],[0],[v[0]], [v[1]], color=colors[i], angles='xy', scale_units='xy', scale=1, alpha=1)
    except:
      plt.quiver([0],[0],[v[0]], [v[1]], color=colors, angles='xy', scale_units='xy', scale=1, alpha=1)
  
  plt.axvline(x=0,c='gray',zorder=0) # zorder for background
  plt.axhline(y=0,c='gray',zorder=0)

  plt.xlim(xlim,ylim)
  plt.ylim(xlim,ylim)

  plt.title(title)
  plt.grid()
  plt.show()
```

```
i = np.array([1,0]) # vectores unitarios
j = np.array([0,1])

draw_vectors([i,j],['red','blue','green'], title='vectores unitarios o base del sistema de coordenadas')
```

```
i = 3*i # vectores escalados
j = 2*j
v = i + j  # combinacion lineal : array([3, 2])

draw_vectors([i,j,v],['red','blue','green'], title='combinacion lineal')
```

**vectores linearmente independientes:** se le dice a un vector linealmente independiente, cuando:

* agrega dimensionalidad, ejemplo claro(i,j,k) canonicos o el espacio generado en el ejemplo anterior con i y j.
* cuando no es una combinacion lineal de otro.
* cuando no es vector nulo(0,0,0).
* si al usar la descomposicion gaussiana tiene una forma escalonada.
* cuando no es identico a otro o no es una vesion escalada del otro.
* cuando no esta alineado con otro vector([1,0],[1,0]).

caso contrario, es un **vector linealmente dependiente**.

por definicion un vector es linealmente independiente cuando:

* $a*i + b*j + c*k = 0: a=b=c=0$

o
* $a*i + b*j ≠ k$


```
v = np.array([1,1])
w = np.array([-1,-1]) # vector linearmente dependiente

# lo comprime todo en una linea

# combinaciones lineales posibles en el rango de -5 a 5
space = np.array([j*v + i*w for i in np.arange(-5,6) for j in np.arange(-5,6)])

draw_vectors(space, xlim=-6, ylim=6, title='Linearmente Dependiente')
```

```
v = np.array([1,0])
w = np.array([0,1])

# combinaciones lineales posibles en el rango de -5 a 5
space = np.array([j*v + i*w for i in np.arange(-5,6) for j in np.arange(-5,6)])

draw_vectors(space, xlim=-6, ylim=6, title='Linearmente Independiente')
```

```
v = np.array([1,1])  # diferente base de sistema de coordenadas
w = np.array([1,-1])

# combinaciones lineales posibles en el rango de -5 a 5
space = np.array([j*v + i*w for i in np.arange(-5,6) for j in np.arange(-5,6)])

draw_vectors(space, xlim=-6, ylim=6, title='Linearmente Independiente',colors='black')
```

**Transformaciones Lineales:** 

como dice el nombre transforma, le damos un vector (V) y nos devuelve otro (W), tipo una funcion, donde nuestro vector de entrada se **desplaza** hacia nuestro vector de salida o transformado, transformado quiere decir movimiento, desplazamiento, traslación. Osea nos devuelve un vector desplazado o trasladado, haciendo que el espacio se transforme.

![](https://compilandoconocimiento.files.wordpress.com/2017/01/plano.gif?w=347&h=195)

e.g:
como vimos en el tema anterior, el ultimo ejemplo donde el espacio se transforma [[1,0],[0,1]] -> [[1,1],[1,-1]].

**Reglas para que sean Transformaciones Lineales:**

* Las rectas siguen siendo rectas.
* El punto de Origen se mantiene fijo.
* Las diagonales deben ser rectas.
* Las lineas de la cuadricula deben permanecer equidistantes y paralelas.

Un ejemplo más de este sería el vector v = [-1,2] que vimos anteriormente, aplicando la formula en las bases tranformadas representadas como matrix [[1,3][-2,0]] nos daria como resultado [-1(1) + 2(3), -1(-2) + 2(0)] = [5, 2].

```
# MULTIPLICACION MATRICIAL:

# | a c |  | x | ___   | a |    | c | ___ | x(a) + y(c) | 
# | b d |  | y | ───  x| b | + y| d | ─── | x(b) + y(d) |

# siendo equivalente a:
# X veces a y b, y Y veces c y d.

# v  = [x, y] = [-1, 2]

# |  1 3 |  | -1 | ___    |  1 |    | 3 | ___ | -1(1)  + 2(3) | ___ | 5 |
# | -2 0 |  |  2 | ───  -1| -2 | + 2| 0 | ─── | -1(-2) + 2(0) | ─── | 2 |
```

siendo el vector v = ([-1,2]) representado en la nueva base del sistema de coordenadas, como 'X' veces 'i' y 'Y' veces 'j', siendo 'i' y 'j' la nueva base de sistema de coordenadas.

```
Matrix = np.array([[1,3],[-2,0]]) # Matrix 2x2 or the representation of [i,j]^T

# Matrix inner product with [-1,2] or v

new_v = Matrix.dot(v) # vector transformado [5,2]^T

draw_vectors([i,j,new_v],['orange','green','gold'],title='transformacion lienal',xlim=-2.5,ylim=5.5)
```

Recordar que el vector [5,2] que se visualiza en la base del sistema de coordenadas [[1,0],[0,1]]^T (vectores canonicos) sería la representacion del vector v=([-1,2]) en la nueva base del sistema de coordenadas transformado [[1,-2],[3,0]]^T.

```
space = np.array([Matrix.dot(np.array([i,j])) for i in np.arange(-5,6) for j in np.arange(-5,6)])

draw_vectors(space, xlim=-12, ylim=12, title='Subespacio de la nueva Transformacion Lineal',colors='blue')
```

```
# vesion 2

import numpy as np
import matplotlib.pyplot as plt

def draw_vectors(vectors, colors='red', xxlim=-1, xylim=5, yxlim=-1, yylim=5, title=''):
  for i,v in enumerate(vectors):
    try:
      plt.quiver([0],[0],[v[0]], [v[1]], color=colors[i], angles='xy', scale_units='xy', scale=1, alpha=1)
    except:
      plt.quiver([0],[0],[v[0]], [v[1]], color=colors, angles='xy', scale_units='xy', scale=1, alpha=1)
  
  plt.axvline(x=0,c='gray',zorder=0) # zorder for background
  plt.axhline(y=0,c='gray',zorder=0)

  plt.xlim(xxlim,xylim)
  plt.ylim(yxlim,yylim)

  plt.title(title)
  # plt.xticks(np.arange(-10,11))
  # plt.yticks(np.arange(-10,11))
  plt.grid()
  plt.show()
```

**y para que nos sirve esto ?**

  pues como vimos antes realizabamos las transformaciones lineales paso a paso, osea para que un vector este en un espacio rotado 90° y despues este en uno inclinado teniamos que hacer dos transformaciones.

e.g:

```
# i = orange & j = green & v = red

i, j = np.array([1, 0]), np.array([0, 1]) # canocical vector

v = np.array([3, 2]) # vector

draw_vectors([i,j,v] , colors = ['orange','green', 'red'],title='normal')
```

```
i2, j2 = np.array([0, 1]), np.array([-1, 0]) # new base or rotation 90°

i = i[0]*i2 + i[1]*j2
j = j[0]*i2 + j[1]*j2

v2 = v[0]*i + v[1]*j

draw_vectors([i,j,v2], colors = ['orange','green','red'],title='first transformation', xxlim=-2.5, xylim=1, yxlim=-1, yylim=3.5)
```

```
i3, j3 = np.array([1, 0]), np.array([1, 1]) # new base or shear

i = i[0]*i3 + i[1]*j3
j = j[0]*i3 + j[1]*j3

v3 = v2[0]*i3 + v2[1]*j3

draw_vectors([i,j,v3], colors = ['orange','green','red'],title='second transformation', xxlim=-1.5, xylim=2, yxlim=-1, yylim=3.5)
```

como vimos hicimos 2 transformaciones, para ahorrarnos ese trabajo nosotros podemos hacer una composicion de matrices o la composicion de transformaciones, multiplicando la primera tranformacion(rotate 90°) por la segunda(shear o inclinacion) obteniendo una nueva matriz, siendo la "Composición" de una rotación e inclinación.

![](https://compilandoconocimiento.files.wordpress.com/2017/01/captura-de-pantalla-2017-02-26-a-las-6-49-53-p-m.png?w=788)


**Importante:** Recordar las propiedades de las matrices cuando hay un producto entre ellas. Son asociativas, distributivas, identidad, pero no conmutativas.

Tambien como observacion, la propiedad "Asociativa" se aplica en la imagen anterior de composicion de matrices. Donde $M1(M2*V) = (M1*M2)V$.

simplificando el trabajo anterior de graficas por codigo sería:


```
# composition of rotate 90° and shear(version 1):

rotate = np.array([i2, j2]).T
shear = np.array([i3, j3]).T

print(shear)
print(rotate)

matrix = shear.dot(rotate)

print(matrix)

new_v = matrix.dot(v)

draw_vectors([matrix[:,0],matrix[:,1],new_v], colors = ['orange','green','red'],title='composition of rotate 90° and shear', 
             xxlim=-1.5, xylim=2, yxlim=-1, yylim=3.5)
```

```
# composition of rotate 90° and shear(version 2):

i, j = np.array([1, 1]), np.array([-1, 0])

new_v = v[0]*i + v[1]*j

draw_vectors([i,j,new_v], colors = ['orange','green','red'],title='composition of rotate 90° and shear', 
             xxlim=-1.5, xylim=2, yxlim=-1, yylim=3.5)
```

**Conmutatividad:**
como explicamos antes, en las propiedades de multiplicacion de matrices puede existir "Asociatividad", "Distribuidad", "Identidad", pero no "Conmutatividad". e.g:

```
import matplotlib.pyplot as plt
import numpy as np
```

```
v = np.array([2,3])

# shear(rotate*v)
first_path = { 'normal': np.array([[1,0],[0,1]]), 'rotate 90°': np.array([[0,-1],[1,0]]), 'shear': np.array([[1,1],[0,1]]).dot(np.array([[0,-1],[1,0]]))}

# rotate(shear*v)
second_path = {'normal': np.array([[1,0],[0,1]]), 'shear': np.array([[1,0],[1,1]]), 'rotate 90°': np.array([[0,-1],[1,0]]).dot(np.array([[1,1],[0,1]]))}
```

```
fig, axs = plt.subplots(2, 3, sharex=True, sharey=True, figsize=(12,8)) # (rows, columns, figsize = (width, height))

colors = ['red','blue']
# v = vector, g = rotate, f = shear

for i, elements in enumerate([first_path, second_path]):
  cade = ''
  for j, element in enumerate(elements):
    x = elements[element]
    for c,e in enumerate([x[:,0], x[:,1]]):
      axs[i,j].quiver([0], [0], [e[0]], [e[1]], angles='xy', scale_units='xy', scale=1, alpha=1, color = colors[c])

    if element =='rotate 90°':
      cade += '(G)'
    elif element =='shear':
      cade +='(F)'
    else:
      cade +='(v)'

    axs[i,j].set_title(element)
    axs[i,j].set_ylabel(cade)

    plt.xlim(-1.5,1.5)
    plt.ylim(-1.5,1.5)
    axs[i,j].axvline(x=0,c='gray',zorder=0)
    axs[i,j].axhline(y=0,c='gray',zorder=0)
    axs[i,j].grid()


plt.tight_layout() # more extent
plt.show()
```

**En Conclusion:**

$F(G(v)) ≠ G(F(v))$, donde $F(x)$ es igual a la inclinacion o shear, $G(x)$ a la rotacion de 90°, y $"v"$ a la base de los vectores canonicos. Es diferente porque según la propiedad conmutativa de multiplicacion de matrices $AB ≠ BA$.

Decir "$L: v -> W$" es equivalente a decir "$L(v)$".

$$Richard-Hamming$$

**Determinantes:**

Los determinantes calculan el **"área"(2D)**, **"Volumen"**(3D), etc. de cierto objeto o matriz, donde como uno  de los principales usos es determinar si nuestro espacio despues de una transformacion lineal crece(aumento, expandio) o decrece(redujo, disminuyo, comprimio).O sea vendría a ser la razon de cambio de nuestro espacio despues de una transformacion lineal, tambien se le conoce como determinante de una tranformacion.

calcular el determinante en 2D:

![](https://compilandoconocimiento.files.wordpress.com/2017/01/16.png?w=429&h=162)

Ejemplos:

```
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np

base = np.array([[1,0],[0,1]])

determinant = np.linalg.det(base)


fig, ax = plt.subplots()

ax.add_patch(
     patches.Rectangle(
        (0.03, 0.03),
        1,1,
        facecolor = 'gold',
        fill=True
     ) )

ax.quiver([0],[0],[1],[0],color='green', angles='xy', scale_units='xy', scale=1, alpha=1)
ax.quiver([0],[0],[0],[1],color='red', angles='xy', scale_units='xy', scale=1, alpha=1)

ax.axhline(y=0,c='gray',zorder=0)
ax.axvline(x=0,c='gray',zorder=0)

plt.xlim(-0.5,4)
plt.ylim(-0.5,4)

plt.title('The determinant is: '+str(determinant))

plt.grid()
plt.show()
```

Como vimos, el área de una matriz canonica en 2d es 1, y eso aplica para cada cuadrado en la cuadricula.

En los siguiente ejemplos donde la combinacion es $"v = 1*i + 1*j"$ o sea la matriz base en si, esta regla tambien sera valida, siempre y cuando sea una transformacion lineal, donde las lineas de la cuadricula sean paralelas y equidistantes, o en general cumpliendo las reglas de transformaciones lineales.

<h1>Determinante igual a Cero:</h1>

Si el determinante es igual a $"0"$ significa que no tiene area, o volumen, dependiendo si es 2D, 3D, etc. ya que al ser un vector linearmente dependiente se reduce la dimensionalidad, donde en 2D se convierte en una linea o punto; y en 3D en un plano, line, o punto.

![](https://compilandoconocimiento.files.wordpress.com/2017/01/14.gif?w=396)


<h1>Determinante Negativo:</h1>

Si el determinante es **"negativo"**, significa que la transformacion giro el plano, osea que el espacio se invierte. Ojo el resultado absoluto es el área.

![](https://compilandoconocimiento.files.wordpress.com/2017/01/15.gif?w=349)

y por qué al girar el plano el determinante se vuelve negativo?

porque al momento de girar el plano el vector $"i"$ se aserca al vector $"j"$ poco a poco comprimiendo las áreas hasta llegar a una linea donde el vector $"i"$ ya no aporta una un valor como dimension siendo linearmente dependiente y el determinante siendo igual a $"cero"$, y por lo cual el área será tambien cero;  despues si el vector $"i"$ se sigue desplazando, el determiante que antes era 0 comenzara a decrecer convirtiendose en negativo, el área será el valor absoluto del determinante.
